#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1
# your answer here
employee.groupby('Department').agg(np.mean)[['Salary']]
```

```python
# Method 2
# your answer here

employee.pivot_table(index='Department',values='Salary')
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
# your answer here

employee.pivot_table(columns=['Department'], index=['Title'], values=['Salary'])
```

#### We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table.
*Hint: Use Pandas doc(see References in README.md) to explore the pivot_table() arguments*

```python
# your answer here
round(employee.pivot_table(index=['Title']),2)
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
# your answer here
employee.pivot_table(columns='Department', index=['Title'], aggfunc={'Salary':np.mean,'Title':'count'}).rename(columns={'Title':'Number of Employees'})
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
# your answer here
round(employee.pivot_table(index='Title', columns='Department', aggfunc={'Salary':[np.mean,np.median,max,np.std], 'Title':'count'}).rename(columns={'Title':'Number of Employees'}))
```

There are missing values for standard deviation when there is only one employee from which to get the mean and missing values for the rest of when there is no record that .

#### Based on your comments, fill the missing values with an appropriate value.

```python
# your answer here

round(employee.pivot_table(index='Title', columns='Department', fill_value=0, aggfunc={'Salary':[np.mean,np.median,max,np.std], 'Title':'count'}).rename(columns={'Title':'Number of Employees'}))
```

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
# your answer here
employee.pivot_table(index=['Department','Title'], aggfunc={ 'Title':'count','Salary':sum}).rename(columns={'Salary':'Salary Total','Title':'Number of Employees'})
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
# your answer here

employee.pivot_table(index=['Department','Title'], aggfunc={'Salary':np.mean, 'Years':sum}).rename(columns={'Salary':'Mean Salary','Years':'Total Years'})
```
